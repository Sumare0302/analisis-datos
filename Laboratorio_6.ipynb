```python
import folium
from folium.plugins import MiniMap
from branca.element import Figure

popuptext = '<b>Estadio Alejandro Morera Soto</b>'
popuptext2 = '<b>Escuela de informatica</b>'

my_map = folium.Map(location=(10.021667,-84.208889),zoom_start=16,tiles='stamenterrain')


folium.Marker(location=(10.021667,-84.208889),popup=popuptext,icon=folium.Icon(color="orange",icon="ok-sing")).add_to(my_map)
folium.Marker(location=(9.9704321,-84.1303143),popup=popuptext2,icon=folium.Icon(color="blue",icon="ok-sing")).add_to(my_map)

folium.Circle(location=[10.021667,-84.208889], color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltip="Estadio Alejandro Morera Soto").add_to(my_map)
folium.Circle(location=[9.9704321,-84.1303143], color="red", weight=6, fill_color="red", fill_opacity=0.5, tooltip="Escuela de informatica").add_to(my_map)

minimap = MiniMap(tile_layer='stamenterrain',position='bottomright')


my_map.add_child(minimap)

my_map.save('C:folium.hmtl')

my_map
```
